```python
import urllib.request
import re
import datetime
import json
```

```python
find_text = re.compile(r'<a href=".*article-.*title="">(.*?)</a>')
```

```python
text_dict = []
```

```python
def get_date_list(start,end):
    date_list= []
    date = datetime.datetime.strptime(start,'%Y%m%d')
    end = datetime.datetime.strptime(end,'%Y%m%d')
    while date <= end:
        date_list.append(date.strftime('%Y%m%d'))
        date = date + datetime.timedelta(1)
    return date_list
```

```python
def askURL(url):
    headers = {
        "accept": "*/*",
        "accept-language": "zh-CN,zh;q=0.9,en;q=0.8,en-GB;q=0.7,en-US;q=0.6",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36 Edg/99.0.1150.30"
    }
    req = urllib.request.Request(url=url, headers=headers)
    html = ""
    try:
        response = urllib.request.urlopen(req)
        html = response.read().decode('utf-8')
    except:
        html = response.read()
    return html
```

```python
def getData(html, date):
    text_list = re.findall(find_text, html)
    for text in text_list:
        text_dict.append({"date": date, "content": text})
```

```python
date_list = get_date_list("20200101", "20210531")
for date in date_list:
    url = "https://www.dailymail.co.uk/home/sitemaparchive/day_" + date + ".html"
    html = askURL(url)
    getData(html, date)
    print("---------已收集至: ", date, "    当前新闻数量：", len(text_dict))
```

```python
with open('../data/news.json', 'a+') as outfile:
    json.dump(text_dict, outfile)
```
